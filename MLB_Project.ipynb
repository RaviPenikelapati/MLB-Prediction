#### **2.2 Data Analysis**
Supervised machine learning was used in the predictive model. Our target variable for all our Regression models is Win% (since we want to be able to predict the success rate, our models will use regression rather than just classification). 

Regression algorithms are trained using percentage split method. They are hyperparameter-tuned and tested using the test set. We used R-squared values and Mean-Squared Error (MSE) as evaluation metrics. 

The following are the Regression models that we used:

* Linear Regression :
  * Fits a linear model with coefficients to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.
  * Since all our features are modeled in such a way that they are in approximately the same range and have similar distributions, a simple linear model could do a good job of generalizing the fit.
  * Features: All.   Target: Win%

* Ridge Regressor:
 * Addresses some of the problems of Ordinary Least Squares by imposing a penalty on the size of the coefficients with L2 regularization.
 * May give slightly better performance than Linear Regressor because it will impose a penalty on the outlier data
 * Features: All.   Target: Win%

* Lasso Regressor:
 * A linear model that estimates sparse coefficients with L1 regularization. Coerces some feature coefficients to zero (auto selection)
 * Since some of the less important features are dropped, might be better able to generalize from the most important features. But we do not expect the best performance before hyperparameter tuning.
 * Features: All.   Target: Win%


* k-Nearest Neighbor Regressor:
 * Regression based on k-nearest neighbors. The target is predicted by local interpolation of the targets associated of the nearest neighbors in the training set.
 * May not perform very well on our dataset because a lot of the data is overlapping since all features are normalized (by the way the stats are calculated) to be in the same range.
 * Features: All.   Target: Win%

* Support Vector Machine Regressor:
 * Good for multidimentional data. Tries to draw decision planes based on decision boundaries in a multidimensional space.
 * We are expecting SVM to perform quite well since we are working with such a large multidimensional data, but it may not be very good before hyperparameter tuning.
 * Features: All.   Target: Win%







```python
import pandas as pd

#Set starting and ending year from available csv data files
start_year = 1986  
end_year = 2021

batting_stats_list = []
pitching_stats_list = []
record_stats_list = []

def add_year(df_list, year):
  '''Adds year column to given list of dataframes'''
  for df in df_list:
    df['Year']=year

def fillMLBData(start, end):
  '''Fills dataframes of all MLB team stats b/w start and end years'''
  for year in range(start,end+1):
    y = str(year)

    #Importing Batting, Pitching, and Win/Loss Record CSVs hosted on github
    batting_df = pd.read_csv('https://raw.githubusercontent.com/jfuiman2000/DS3000-Project-DataSet/main/' + y+'BattingStats.csv')
    pitching_df = pd.read_csv('https://raw.githubusercontent.com/jfuiman2000/DS3000-Project-DataSet/main/' + y+'PitchingStats.csv')
    record_df = pd.read_csv('https://raw.githubusercontent.com/jfuiman2000/DS3000-Project-DataSet/main/' + y+'Record.csv')

    #Adding year column (Wrangling but convenient to do here)
    add_year([batting_df, pitching_df, record_df], year)  

    #Creating a list of dataframes corresponding to years
    batting_stats_list.append(batting_df)
    pitching_stats_list.append(pitching_df)
    record_stats_list.append(record_df)

fillMLBData(start_year,end_year)
```

## **3 RESULTS**

#### **3.1 Data Wrangling**

* **Data cleaning**:
 * We are only interested in specific columns from the batting, pitching and records dataframes. Therefore, we **dropped** the rest of the columns from each dataframe. 
 * **Concatenated** lists of dataframes (corresponding to years between 1986-2021) into three combined dataframes for batting, pitching, and records. 
 * **Renamed** columns to remove inconsistencies and duplicate names. 
 * Created a new column called Win% which is calculated using the W (wins) and L (Losses) columns in the Records dataframe. This will be our target column.
 * **Merged** the three dataframes into one single dataframe. We were able to do this because we had team and year columns which were used to join all columns into a single dataframe.

```python
#Listing required features from each dataframe
batting_features = ['Team', 'BB%+', 'K%+', 'AVG+', 'OBP+', 'SLG+', 'ISO+', 'Year']
pitching_features = features = ['Team', 'BB%+', 'K%+', 'AVG+', 'HR/9+', 'WHIP+', 'ERA-', 'FIP-', 'Year']
record_features = ['Team', 'W', 'L', 'Year']

batting_list=[]
pitching_list=[]
record_list=[]

#Dropping unnecesary columns from all dataframes
for batting_df, pitching_df, record_df in zip(batting_stats_list, pitching_stats_list, record_stats_list):
  batting_list.append(batting_df[batting_features])
  pitching_list.append(pitching_df[pitching_features])
  record_list.append(record_df[record_features])

#Concatenating the lists of dataframes into batting, pitching, and record dataframes
#Sorting dataframes by year and team
#Resetting indices
batting_df = pd.concat(batting_list).sort_values(by=['Year', 'Team']).reset_index(drop=True)
pitching_df = pd.concat(pitching_list).sort_values(by=['Year', 'Team']).reset_index(drop=True)
record_df = pd.concat(record_list).sort_values(by=['Year', 'Team']).reset_index(drop=True)

#Rename columns with same name in batting and pitching dataframes
batting_df.rename(columns={'BB%+':'B_BB%+', 'K%+':'B_K%+', 'AVG+':'B_AVG+'}, inplace=True)
pitching_df.rename(columns={'BB%+':'P_BB%+', 'K%+':'P_K%+', 'AVG+':'P_AVG+'}, inplace=True)

#Calculating Win Percentage from Wins/Losses and adding to new column (win_ratio = (wins*100)/(wins+losses))
record_df['Win%'] = (record_df['W'] * 100) / (record_df['L'] + record_df['W'])
record_df = record_df[['Team', 'Year', 'Win%']]

#Merging batting, pitching, and record dataframes
MLB_df = pd.merge(pd.merge(batting_df,pitching_df), record_df)

MLB_df.head()
```

* **Extracting feature and target values**:
 * We are not including the year as one of the features (the year information does not add any useful information in the prediction of success. Therefore, we will drop it using the to_drop parameter).
 * The Win% column will be our target variable to indicate the success of a team in MLB. 


```python
def features_target(df, target, to_drop=[], to_select=[]):
    '''Returns the target variable and features with dropped values'''

    #Selects and sets the required features
    if to_select:
        features = df[to_select]
    else:
        to_drop.append(target)
        features = df.drop(to_drop, axis=1)
    
    #Win% is the column name of the target variable
    target = MLB_df[target]

    return features, target 
```

* **Preprocessing variables**:
 * We are not normalizing our dataframe because the range of the feature variables is not very different. 
 * We used MinMax scaling to scale our feature variables to fit within the range [0-1].

```python
def preprocess(train, test):
    '''Returns scaled feature variables'''
    from sklearn.preprocessing import MinMaxScaler

    #Fitting MinMax Scaler with our feature variables' values.
    scaler = MinMaxScaler()
    scaler.fit(train) 
    
    #Scaling feature variables [0 - 1]
    xtrain_scaled = scaler.transform(train) 
    xtest_scaled = scaler.transform(test)

    return xtrain_scaled, xtest_scaled 
```

* **Feature Extraction**:
 * We felt that it was not very necessary to extract a lot of additional features from existing features. This is because most of the features are already informative on their own and also normalized.
 * The only feature that we felt could be informative is combining FIP and ERA stats because they are correlated. If a pitcher has surrendered a high average on balls in play, his FIP will likely be lower than his ERA. Balls in play are not part of the FIP equation because a pitcher is believed to have limited control over their outcome. 

```python
#DO NOT RUN TWICE

# Control = (FIP- / ERA-) * 100
MLB_df['Control'] = MLB_df['FIP-'] / MLB_df['ERA-'] * 100
# Setting control to int to match the rest of the data
MLB_df['Control'] = MLB_df['Control'].astype(int)

#Swaps control and Win%
cols = list(MLB_df.columns)
a, b = cols.index('Win%'), cols.index('Control')
cols[b], cols[a] = cols[a], cols[b]
MLB_df = MLB_df[cols]

MLB_df.head()
```

#### **3.3 Model Training**

* **Train/Test Split**: We performed a Train-Test Split to Percentage-Split our data set for training our machine learning models and testing to evaluate their performance.



```python
def percentage_split(features_to_split, target_to_split, pre=False):
    '''Returns training and testing splits'''
    from sklearn.model_selection import train_test_split

    #Splits into training and testing variables
    X_train, X_test, y_train, y_test = train_test_split(features_to_split, target_to_split, random_state = 3000)
    
    #Preprocess feature variables if specified
    if pre:
        X_train, X_test = preprocess(X_train, X_test)
    
    return X_train, X_test, y_train, y_test
```

* **Fitting ML Models**: We fit the training data to the most commonly used linear regression models (Linear, Ridge, Lasso, k-Nearest Neighbor, Support Vector Machine). In order to make the fitting process simpler we created a dictionary of Regression Models that we wanted to use

```python
#All required imports
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
import numpy as np
from sklearn.model_selection import GridSearchCV
```

```python
def regression(select=14, preprocess=False, tuned=False, verbose=True):
  '''Returns results from test set after fitting regression models'''
  #Breaking dataset down to features and target
  features, target = features_target(MLB_df, 'Win%', to_drop=['Team','Year'])
  #Splitting into test and train set (optional preprocessing)
  xtrain, xtest, ytrain, ytest = percentage_split(features, target, pre=preprocess)
  models, trains, tests, mtrains, mtests=[],[],[],[],[]

  #Optional feature selection using RFE
  if not select==14:
    xtrain, xtest = select_RFE(select, xtrain, xtest, ytrain, features.columns, show_features=verbose)

  #Optional model tuning
  if tuned:
    #Tuned set of Regression Models (Note: contains GridSearchCV objects which behave like Models)
    estimators= tuned_models(xtrain, xtest, ytrain, ytest, show_params=verbose)
  else:
    #Default set of Regression Models
    estimators = {'Linear':LinearRegression(), 'Ridge':Ridge(), 'Lasso':Lasso(alpha=0.001), 'kNN':KNeighborsRegressor(), 'SVM':SVR(max_iter = 1000000)}

  #Fits each model with training data and stores performance on test set
  for model_name, model in estimators.items():
    model.fit(X = xtrain, y = ytrain)

    #R-squared training scores
    train_score = r2_score(ytrain, model.predict(xtrain)) * 100
    test_score = r2_score(ytest,  model.predict(xtest)) * 100

    #Mean Squared Error values
    MSE_train = mean_squared_error(ytrain, model.predict(xtrain)) 
    MSE_test = mean_squared_error(ytest, model.predict(xtest)) 

    models.append(model_name)
    trains.append(round(train_score, 2))
    tests.append(round(test_score,2))
    mtrains.append(round(MSE_train,2))
    mtests.append(round(MSE_test,2))
  
  results={'Model': models, 'Train': trains, 'Test': tests, 'MSETrain': mtrains, 'MSETest': mtests}
  return results
```

* **Feature Selection**: We decided to use Recursive Feature Elimination (RFE) to select the five most important features. We chose RFE because we felt that it would recursively select the most important features based on the Decision Tree Regressor model. 

```python
def select_RFE(size, train, test, y, col, show_features=True):
  from sklearn.feature_selection import RFE
  from sklearn.tree import DecisionTreeRegressor
  
  #RFE selector that selects the most important features according to Decision Tree Regressor
  selector = RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select = size)
  selector.fit(train,y)
  
  X_train_selected = selector.transform(train)
  X_test_selected = selector.transform(test)

  #Prints the features selected by RFE
  selected_features= [feature for feature,selected in zip(col, selector.get_support()) if selected]

  if show_features:
    print("Features selected after RFE:")
    for feature in selected_features:
      print(feature)

  return X_train_selected, X_test_selected
```

**Results Evaluation**: kNN and SVM have slight improvements after feature selection. Linear, Ridge, and Lasso are slightly worse than using all features. 

One interesting observation is that kNN algorithm seems to have significantly higher Score and lower MSE on the training set. It might be because kNN is able to fit the training set data better than the rest of the models.

Also, we compared multiple values for the number of selected features. From 1 feature to 5 there are significant leaps in performance across all models. However, after the top 5 feature variables, the performance improvements slow down significantly for all models (reduction in performace for kNN). This suggests to us that the five selected features are vital to success in the MLB according to RFE and model testing.

#### **3.4 Model Optimization**
We used GridSearchCV to hyperparameter tune our Regression Models. We felt that it made sense to try to tune all our models because the worse performing models (kNN and SVM) looked like they were overfit and underfit respectively (as detailed in results evaluations). Therefore they might have more scope for improvement. 

```python
#Grid search values are intentionally of smaller range (to reduce loading time on method calls)
def grid_search_Ridge(xtrain, ytrain):
    '''Returns tuned Ridge Regressor'''
    param_grid = {"alpha": [x/1000 for x in range(130,140)]}  
    grid_search = GridSearchCV(Ridge(), param_grid, cv=5)
    grid_search.fit(X=xtrain, y=ytrain) 
    return grid_search

def grid_search_Lasso(xtrain, ytrain):
    '''Returns tuned Lasso Regressor'''
    param_grid = {"alpha":[x/10000 for x in range(7,10)]}
    grid_search = GridSearchCV(Lasso(tol=0.0005), param_grid, cv=5)
    grid_search.fit(X=xtrain, y=ytrain) 
    return grid_search

def grid_search_kNN(xtrain, ytrain):
    '''Returns tuned kNN Regressor'''
    param_grid = {"n_neighbors":list(range(15,20)), "metric":["euclidean", "manhattan", "minkowski"]} 
    grid_search = GridSearchCV(KNeighborsRegressor(), param_grid, cv=5)
    grid_search.fit(X=xtrain, y=ytrain) 
    return grid_search

def grid_search_SVM(xtrain, ytrain):
    '''Returns tuned SVM Regressor'''    
    param_grid = {"C":list(range(97,103)), 'gamma': [1, 0.1, 0.01]} 
    grid_search = GridSearchCV(SVR(max_iter = 1000000), param_grid, cv=5)
    grid_search.fit(X=xtrain, y=ytrain) 
    return grid_search

#Note: Returns GridSearchCV objects which behave like Models
def tuned_models(xtrain, xtest, ytrain, ytest, show_params=True):
  '''Returns a dictionary of tuned models and prints best parameters in each'''
  tuned_estimators = {'Linear':LinearRegression(), 'Ridge':grid_search_Ridge(xtrain, ytrain), 'Lasso':grid_search_Lasso(xtrain, ytrain), 'kNN':grid_search_kNN(xtrain, ytrain), 'SVM':grid_search_SVM(xtrain, ytrain)}
  
  if show_params:
    for model_name, model in tuned_estimators.items():
      if not model_name=='Linear':
        print(model_name+" best parameters:  ", model.best_params_)

  return tuned_estimators
```
